The Double Deep-Q-Network procedure ddqn performs the double loop. The external loop (by episodes) is executed until the number of episodes reaches the maximal number n_episodes=2000 or the completion criteria is met. The environment env is reset with the parameter train_mode=True. For the completion criteria, we check

                    np.mean(scores_window)≥13,

where scores_window is a deque implementing a shifting window of length ≤100. The element scores_window[i] contains the score achieved by the algorithm on episode i.

In the internal loop, ddqn gets the current action from the agent. By this action, ddqn receives the next_state, reward, and done signal from the Unity environment env. Then, the agent accepts parameters (state, action, reward, next_state, done) in the next training step. The variable score accumulates the obtained rewards.

````python
import time
import math
import datetime
import torch
from ddqn_agent import ddqn_Agent  # Import DDQN agent
from collections import deque
import matplotlib.pyplot as plt
import numpy as np  # Import numpy for np.mean
%matplotlib inline 

def ddqn(n_episodes=2000, eps_start=.99, eps_end=0.01, eps_decay = .996, train_numb = 0):
    """Deep Double Q-Learning.
    
    Params
    ======
        n_episodes (int): maximum number of training episodes
        eps_start (float): starting value of epsilon, for epsilon-greedy action selection
        eps_end (float): minimum value of epsilon
        eps_decay (float): multiplicative factor (per episode) for decreasing epsilon
    """
    start_time = time.time()                               # start time 
    scores = []                                            # list of scores for each episode
    scores_window = deque(maxlen=100)                      # last 100 scores
    eps = eps_start                                        # initialize epsilon

    # Initialize the environment and agent
    # Assuming env and brain_name are defined elsewhere in your code
    # If not, you need to define them before using them.
    agent = ddqn_Agent(state_size=37, action_size=4, seed=0)  # Initialize DDQN agent

    for i_episode in range(1, n_episodes+1):               # loop by episodes
        env_info = env.reset(train_mode = True)[brain_name]
        state = env_info.vector_observations[0]            # get the current state 
        score = 0                                          # reset the score counter
        done = False                                       # are we done yet?
        while not done:                                    # internal loop in the episode
            action = agent.act(state,eps)                  # next action from the agent 
            action = int(action)                           # cast to int
            env_info = env.step(action)[brain_name]        # send the action to the environment
            next_state = env_info.vector_observations[0]   # get the next state
            reward = env_info.rewards[0]                   # get the reward
            done = env_info.local_done[0]                  # done is true if episode has finished
            agent.step(state,action,reward,next_state, done) # next learning step by state and reward
            score += reward                                # count our rewards
            state = next_state                             # update state
            if done:                                       # done ?
                break                                      # save score
        scores_window.append(score)                        # save score in the deque with 100 or less elements
        scores.append(score)
        
        eps = max(eps_end,eps_decay*eps)                   # make epsilon a bit smaller
        
        count = 0                                          # how many times we've reached 13
        for j in range(len(scores_window)):                
            if scores_window[j] >= 13:
                count+=1
                
        elapsed = datetime.timedelta(seconds = time.time()-start_time)  # elapsed time
        
        print('\rEpisode: {}, elapsed: {}, Avg.Score: {:.2f},  score {}, How many scores >= 13: {}, eps.: {:.2f}'. \
            format(i_episode, elapsed, np.mean(scores_window), score, count, eps), end="")
        
        if np.mean(scores_window) >=13:  # check completion criteria.
            print("\n terminating at episode :", i_episode, "ave reward reached +13 over 100 episodes")
            break
            
    torch.save(agent.qnetwork_local.state_dict(), 'weights_ddqn_'+str(train_numb)+'.trn') # save the weights into the file 
    return scores, i_episode
````

### 6. Testing session and the average score

We load weights by the function _torch.save_ of the package **PyTorch** from one of the saved weights files:

    file_weights = 'weights_'+str(train_n)+'.trn'
    agent.qnetwork_local.load_state_dict(torch.load(file_weights))

We run testing for each file _weights_\__i.trn_, where i = 0,1,2,...
The testing procedure is implemented by the function **checkWeights**.
The function **checkWeights** constructs the **agent** with 
the parameters _fc1_\__units_ and _fc2_\__units_ corresponding to the given weights file.
For each testing session, we run **checkWeights** several times (=6 in this version) 
to get the average score for the given set of parameters.

````python
import torch
from unityagents import UnityEnvironment
import numpy as np
from ddqn_agent import ddqn_Agent  # Ensure this imports your DDQN Agent class

# Load the environment
env = UnityEnvironment(file_name="C:\\Users\\ADMIN\\Deep-Reinforcement-Learning-Algorithms\\Project-1_Navigation-DQN\\Banana_Windows_x86_64\\Banana.x86_64",worker_id=3)

# Get the brain
brain_name = env.brain_names[0]
brain = env.brains[brain_name]

def checkWeights(env, train_n, test, fc1_n, fc2_n, eps_s, episodes):
    agent = ddqn_Agent(state_size=37, action_size=4, seed=17, fc1_units=fc1_n, fc2_units=fc2_n)  
    file_weights = f'weights_ddqn_{train_n}.trn'
    agent.qnetwork_local.load_state_dict(torch.load(file_weights))

    env_info = env.reset(train_mode=False)[brain_name]  # reset the environment
    state = env_info.vector_observations[0]  # get the current state
    score = 0  # initialize the score
    while True:
        action = agent.act(state, 0.05)  # select an action
        action = int(action)
        env_info = env.step(action)[brain_name]  # send the action to the environment
        next_state = env_info.vector_observations[0]  # get the next state
        reward = env_info.rewards[0]  # get the reward
        done = env_info.local_done[0]  # see if episode has finished
        score += reward  # update the score
        state = next_state  # roll over the state to next time step
        if done:  # exit loop if episode finished
            break
    
    # print(f'Train: {train_n}, Test: {test}, Episode: {episodes}, fc1_units: {fc1_n}, fc2_units: {fc2_n}, eps_start: {eps_s}, Score: {score}')
    print('Train: {}, Test: {}, Episode: {}, fc1_units: {}, fc2_units: {}, eps_start: {}, Score: {}'\
          .format(train_n, test, episodes, fc1_n, fc2_n, eps_s, score))
    return score

numb_of_trains = 3  # Number of trained models
list_fc1_units = [64, 64, 64]  # Adjust based on your model architecture
list_fc2_units = [64, 64, 64]
list_eps_start = [0.05, 0.05, 0.05]
list_episodes = [1, 1, 1]

for i in range(numb_of_trains):
    fc1_nodes = list_fc1_units[i]
    fc2_nodes = list_fc2_units[i]
    eps_start = list_eps_start[i]
    episodes = list_episodes[i]
    list_scores = []
    for test in range(6):  
        score = checkWeights(env=env, train_n=i, test=test, fc1_n=fc1_nodes, fc2_n=fc2_nodes, eps_s=eps_start, episodes=episodes)
        list_scores.append(score)
    avg_score = np.mean(list_scores)
    print(f'       Average Score: {avg_score}')
    print('=========================================================')

env.close()
````
